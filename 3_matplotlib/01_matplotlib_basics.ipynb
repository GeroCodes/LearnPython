```python
import numpy as np
import matplotlib.pyplot as plt

#use %matplotlib inline to plot within a notebook
%matplotlib inline

#use plt.show() after every plot-method to plot within an editor
#plt.show()
```

## #1 Plotting

```python
#create a dataset
x = np.linspace(0,5,11)
y = x**2

print(x)
print(y)
```

```python
#IMPORTANT: MATPLOTLIB
#plot #xlabel #ylabel #matplotlib

#regular plotting
plt.plot(x,y, label='label_show')
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.title('Ãœberschrift')
```

## #2 SUBPLOTS

```python
#IMPORTANT: MATPLOTLIB
#subplot #plot #matplotlib

#Subplots: create a variety of plots next to each other
#.sublot(number of rows, number of cols, plot-number)
plt.subplot(2,2,1) #plot upper left
plt.plot(x,y,'r-') #plot red line
plt.subplot(2,2,2)
plt.plot(y,x,'g*') #plot dotted green
plt.subplot(2,2,3)
plt.plot(y,x,'b-')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

## #3a Object-oriented Plotting: plt.figure creates a single figure with more than one axes

```python
#IMPORTANT: MATPLOTLIB
#plt.figure #add_axes #plot #matplotlib

#object 1: create empty figure called af
#determine the size with figsize
fig = plt.figure(figsize=(4,2),dpi=100)

#object 2, 3: add axes
#add_axes([pos left, pos bottom, width, height])
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.3])

#add properties for object 2 "axes1"
axes1.plot(x,y,'r-')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

#add properties for obeject 3 "axes2"
axes2.plot(y,x,'b*')
axes2.set_xlabel('x2')
axes2.set_ylabel('y2')
axes2.set_title('title2')
```

## 3b Object-oriented Plotting: creating multiple figures with plt.subplots

```python
#create subplot-objects
#create empty figure diag and axes with plt.subplot
diag, axes = plt.subplots(nrows=1,ncols=2)

#add properties for axes-objects
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#create empty figure diag and axes with plt.subplot
iter_diag, iter_axes = plt.subplots(nrows=1,ncols=2)

#axes are iterable objects
for ax in iter_axes:
    ax.plot(x,y)
    ax.set_title('title')
    ax.set_ylabel('y-label')

#distancing plots to give y-label some space  
plt.tight_layout()
```
