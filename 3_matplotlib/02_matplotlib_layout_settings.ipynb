# MATPLOTLIB: DATA VISUALIZATION


```python
import numpy as np
import matplotlib.pyplot as plt

#use %matplotlib inline to plot within a notebook
%matplotlib inline

#use plt.show() after every plot-method to plot within an editor
#plt.show()
```

## #1 Layout and Settings

```python
#create a dataset
x = np.linspace(0,5,11)
y = x**2

print(x)
print(y)
```

### Adding a legend, line-settings

```python
#create example for setting layout
diag, axes = plt.subplots(figsize=(10,4))

#editing appereance: lw = linewidth, ls = linestype
axes.plot(y,x, label='label_show', color='red', lw=4, ls='-', marker='+', markersize=14, markeredgewidth=6, markerfacecolor='blue', markeredgecolor='green')

#add legend and define position (alternative: with e.g. loc=1)
axes.legend(bbox_to_anchor=(1,1))

#fix axis
axes.set_ylim(1)

#adding label and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
```

### Edit scale- and axis-parameters

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#scale axis (e.g. log)
axes.set_yscale('log')

#set scientific axis
axes.set_xticks([0, 50, 100, 150])
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
axes.xaxis.set_major_formatter(formatter)

#adding label and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
```

### Edit Grid- and Spine-Settings

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#grid: change settings
axes.grid(True)
axes.grid(color='grey', alpha=0.5, linestyle='dashed', linewidth=0.5)

#splines: hange settings
#also: "top", "left", "right"
axes.spines['bottom'].set_color('red')
axes.spines['bottom'].set_linewidth(2)
```

### Add second y-axis

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#set label axes
axes.set_ylabel(r'area $(m^2)$', fontsize=12, color='grey')

#use iterable objects to determine properties
for label in axes.get_yticklabels():
    label.set_color('blue')

#create a twin and determine properties
axes2 = axes.twinx()
axes2.plot(x,y)
axes2.set_ylabel(r'volume $(m^3)$', fontsize=12, color='grey')

for label in axes2.get_yticklabels():
    label.set_color('red')
```

### Set Text into diagramm

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(x,y)

#add text
#.text(pos left, pos bottom, text, properties)
axes.text(0.15,0.2, r'$y=x^2$', fontsize=20, color='blue')
```

## #5 Save a figure as image-file

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#adjust to make sure the image-file has a proper quality (nothing gets cut off)
diag.subplots_adjust(left=0.5, right=0.9, bottom=0.1, top=0.9)

#save as img-file (plt.subplot = diag, plt.figure = axes)
diag.savefig('image.jpg')
```
