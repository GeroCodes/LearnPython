## How to use Groupby with pandas? 

```python
import pandas as pd
```

```python
#create dataframe
drinks_path = 'http://bit.ly/drinksbycountry'
drinks = pd.read_csv(drinks_path)
```

```python
drinks.head(3)
```

```python
drinks.beer_servings.mean()
```

```python
#IMPORTANT: PANDAS
#agg #groupby #select #explore #pandas

#show multiple KPIs of beer_servings for each continent
drinks.groupby('continent').beer_servings.agg(['count','min',
'max','mean'])
```

```python
#show beer_servings for a particular continent
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
 drinks.groupby('continent').mean()
```

```python
 drinks.groupby('continent').mean().plot(kind='bar')
```
