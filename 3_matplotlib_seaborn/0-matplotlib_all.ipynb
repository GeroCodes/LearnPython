# MATPLOTLIB: DATA VISUALIZATION


```python
import numpy as np
import matplotlib.pyplot as plt

#use %matplotlib inline to plot within a notebook
%matplotlib inline

#use plt.show() after every plot-method to plot within an editor
#plt.show()
```

## #1 Plotting

```python
#create a dataset
x = np.linspace(0,5,11)
y = x**2

print(x)
print(y)
```

```python
#IMPORTANT: MATPLOTLIB
#plot #xlabel #ylabel #matplotlib

#regular plotting
plt.plot(x,y, label='label_show')
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.title('Überschrift')
```

## #2 SUBPLOTS

```python
#IMPORTANT: MATPLOTLIB
#subplot #plot #matplotlib

#Subplots: create a variety of plots next to each other
#.sublot(number of rows, number of cols, plot-number)
plt.subplot(2,2,1) #plot upper left
plt.plot(x,y,'r-') #plot red line
plt.subplot(2,2,2)
plt.plot(y,x,'g*') #plot dotted green
plt.subplot(2,2,3)
plt.plot(y,x,'b-')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

## #3a Object-oriented Plotting: plt.figure creates a single figure with more than one axes

```python
#IMPORTANT: MATPLOTLIB
#plt.figure #add_axes #plot #matplotlib

#object 1: create empty figure called af
#determine the size with figsize
fig = plt.figure(figsize=(4,2),dpi=100)

#object 2, 3: add axes
#add_axes([pos left, pos bottom, width, height])
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.3])

#add properties for object 2 "axes1"
axes1.plot(x,y,'r-')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

#add properties for obeject 3 "axes2"
axes2.plot(y,x,'b*')
axes2.set_xlabel('x2')
axes2.set_ylabel('y2')
axes2.set_title('title2')
```

## 3b Object-oriented Plotting: creating multiple figures with plt.subplots

```python
#create subplot-objects
#create empty figure diag and axes with plt.subplot
diag, axes = plt.subplots(nrows=1,ncols=2)

#add properties for axes-objects
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#create empty figure diag and axes with plt.subplot
iter_diag, iter_axes = plt.subplots(nrows=1,ncols=2)

#axes are iterable objects
for ax in iter_axes:
    ax.plot(x,y)
    ax.set_title('title')
    ax.set_ylabel('y-label')

#distancing plots to give y-label some space  
plt.tight_layout()
```

## #4 Layouts and Settings

### Adding a legend, line-settings

```python
#create example for setting layout
diag, axes = plt.subplots(figsize=(10,4))

#editing appereance: lw = linewidth, ls = linestype
axes.plot(y,x, label='label_show', color='red', lw=4, ls='-', marker='+', markersize=14, markeredgewidth=6, markerfacecolor='blue', markeredgecolor='green')

#add legend and define position (alternative: with e.g. loc=1)
axes.legend(bbox_to_anchor=(1,1))

#fix axis
axes.set_ylim(1)

#adding label and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
```

### Edit scale- and axis-parameters

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#scale axis (e.g. log)
axes.set_yscale('log')

#set scientific axis
axes.set_xticks([0, 50, 100, 150])
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
axes.xaxis.set_major_formatter(formatter)

#adding label and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
```

### Edit Grid- and Spine-Settings

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#grid: change settings
axes.grid(True)
axes.grid(color='grey', alpha=0.5, linestyle='dashed', linewidth=0.5)

#splines: hange settings
#also: "top", "left", "right"
axes.spines['bottom'].set_color('red')
axes.spines['bottom'].set_linewidth(2)
```

### Add second y-axis

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#set label axes
axes.set_ylabel(r'area $(m^2)$', fontsize=12, color='grey')

#use iterable objects to determine properties
for label in axes.get_yticklabels():
    label.set_color('blue')

#create a twin and determine properties
axes2 = axes.twinx()
axes2.plot(x,y)
axes2.set_ylabel(r'volume $(m^3)$', fontsize=12, color='grey')

for label in axes2.get_yticklabels():
    label.set_color('red')
```

### Set Text into diagramm

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(x,y)

#add text
#.text(pos left, pos bottom, text, properties)
axes.text(0.15,0.2, r'$y=x^2$', fontsize=20, color='blue')
```

## #5 Save a figure as image-file

```python
diag, axes = plt.subplots(figsize=(10,4))
axes.plot(y,x)

#adjust to make sure the image-file has a proper quality (nothing gets cut off)
diag.subplots_adjust(left=0.5, right=0.9, bottom=0.1, top=0.9)

#save as img-file (plt.subplot = diag, plt.figure = axes)
diag.savefig('image.jpg')
```

## #6 Advanced concepts for subplots

### Subplot2Grid

```python
#create a figure
fig = plt.figure()

#create axes on the figure
#use colspan and rowspan to determine size
#.subplot2grid((total rows, total cols), (start row, start col), width/height)
ax1 = plt.subplot2grid((3,3),(0,0), colspan=3)
ax2 = plt.subplot2grid((3,3),(1,0), colspan=2)
ax3 = plt.subplot2grid((3,3),(1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0))
ax5 = plt.subplot2grid((3,3),(2,1))
```

### Gridspec

```python
import matplotlib.gridspec as gridspec

#create a figure
fig = plt.figure()

#create gridspec
gs = gridspec.GridSpec(2,3, height_ratios=[2,1], width_ratios=[1,2,1])

for g in gs:
    ax = fig.add_subplot(g)
```
