# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

## #3 Matrixplots

```python
#create matrix like .corr()
tips.corr()
```

```python
#IMPORTANT: SEABORN
#heatmap #correlation #regression #seaborn

#Heatmap: deliver data in a matrix
#often in use with pd.pivot_table()
#visualize correlation with a headmap (also: isnull() to show null-values)
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True,
           linecolor ='white', linewidth=1)
```

```python
#create a pivot-table
flights_pivot=flights.pivot_table(values='passengers', index='month', columns='year')
flights_pivot
```

```python
#show heatmap of pivot-table
sns.heatmap(flights_pivot, cmap='magma', linecolor='white', linewidth=1)
```

```python
#IMPORTANT: SEABORN
#clustermap #correlation #regression #seaborn

#Clustermap: shows a clustered version of the heatmap
#rows and cols are ordered by similarity
sns.clustermap(flights_pivot, cmap='magma')
```

```python
sns.clustermap(flights_pivot, cmap='coolwarm', standard_scale=1)
```
