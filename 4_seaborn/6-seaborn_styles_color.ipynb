# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## #6 Styles and Colors


```python
#set size (hast to be in front of the plot)
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips, palette='deep')

#set background-color
sns.set_style('darkgrid')

#delete frame
sns.despine(left=True)
```

```python
#set size of a plot
sns.lmplot(x='total_bill', y='tip', data=tips, size=2, aspect=4)
```

```python
#set_context to set various customized parameters (e.g. font_scale, line-size)
sns.set_context('paper',font_scale=2)
sns.countplot(x='sex', data=tips, palette='seismic')
```
