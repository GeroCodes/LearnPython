# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## #1 Distribution-Plots

```python
#IMPORTANT: SEABORN
#distplot #distribution #seaborn
#Distplot: visualize distribution
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

```python
#IMPORTANT: SEABORN
#scatter #jointplot #distribution #seaborn

#Scatter-Plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#IMPORTANT: SEABORN
#hex #jointplot #distribution #seaborn

#Hex-Plot: visualize density of a distribution
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#IMPORTANT: SEABORN
#regression #jointplot #distribution #seaborn

#Regressions-Diagramm
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#IMPORTANT: SEABORN
#kde #jointplot #distribution #seaborn

#KDE-Diagramm (Kernel-Density-Estimation)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#IMPORTANT: SEABORN
#pairplot #distribution #bestpractice #seaborn

#Pair-Plot gives a good overview
sns.pairplot(tips)
```

```python
#use pairplot with hue-parameter
sns.pairplot(tips,hue='sex', palette='coolwarm')
```

```python
#IMPORTANT: SEABORN
#rugplot #distribution #seaborn

#Rugplots (density of an univariat distribution)
sns.rugplot(tips['total_bill'])
```

```python
#understanding the rugplot
#visualize two output plots in a single figure
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
