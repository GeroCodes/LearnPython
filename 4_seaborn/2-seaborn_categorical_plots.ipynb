# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## #2 Categorical Plots

```python
#IMPORTANT: SEABORN
#barplot #categoricalplot #seaborn

#Barplot: aggregate categorical features
#set the parameter estimator two decide which KPI to show e.g. np.mean, np.std, np.sum
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#IMPORTANT: SEABORN
#countplot #categoricalplot #seaborn

#Countplot: number of a categorical feature
sns.countplot(x='sex', data=tips)
```

```python
#IMPORTANT: SEABORN
#boxplot #categoricalplot #seaborn

#Boxplots: distribution of categorical features (Whisker-Plot)
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
#change orientation
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
#grouped Boxplot
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='coolwarm')
```

```python
#IMPORTANT: SEABORN
#violinplot #categoricalplot #seaborn

#Violin-Plot: distribution of a categorical feature
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True, palette='Set1')
```

```python
#IMPORTANT: SEABORN
#stripplot #categoricalplot #seaborn

#Strip-Plot: Scatter-Plot with a categorical feature
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='Set1', dodge=True)
```

```python
#IMPORTANT: SEABORN
#swarmplot #categoricalplot #seaborn

#Swarm-Plot: similar to strip-plot
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set2', dodge=True)
```

```python
#combining two categorical plots
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=3)
```

```python
#Factor-Plot: general expression for plots
#set the parameter "kind" to the plot you want to show
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```
