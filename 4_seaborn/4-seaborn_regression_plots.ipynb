# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## #4 Regressionplots

```python
#IMPORTANT: SEABORN
#implot #regression #seaborn

#lmplot: regular regression-plot
#use hue-parameter to group data
#edit markersize with scatter_kws
sns.lmplot(x='total_bill', y='tip',data=tips, hue='sex',
          markers=['o','v'], scatter_kws={'s':50})
```

```python
#IMPORTANT: SEABORN
#implot #grid #multipleplot #regression #seaborn

#create more than a single plot with parameters row and col
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips,
           row='sex', col='time')
```

```python
#change size and aspect
sns.lmplot(x='total_bill', y='tip', data=tips, col='time',
          aspect=0.6, size=7)
```
