# SEABORN: DATA VISUALIZATION

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create example
#delivers standard-datasets e.g. flights, iris, tips,...
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## #5 Grids

### Grids are a general types of plots which have a more customizable character.

### Pairplot: repetition of #1, customizable method

```python
#IMPORTANT: SEABORN
#pairplot #multipleplot #regression #seaborn

#Pairplot (see #1 Distribution Plots)
#Pairplots are a fast and easy version of a PairGrid
#editing of a pairplot is also possible
k=sns.pairplot(iris, hue='species')
k.map_diag(plt.hist, bins=6)
k.map_upper(sns.regplot)
```

### PairGrid: customizable like pairplot-method

```python
#IMPORTANT: SEABORN
#pairgrid #multipleplot #regression #seaborn

#PairGrids are enabled to use different APIs like matplotlib
#PairGrids detect numeric variables automatically
g = sns.PairGrid(iris)
g.map_diag(plt.hist) #diagrams on diag
g.map_upper(plt.scatter) #diagrams above diag
g.map_lower(sns.kdeplot) #diagrams beneath diag
```

### FacetGrid: create customized Grids organized in rows and cols

```python
#IMPORTANT: SEABORN
#facetgrid #multipleplot #regression #seaborn

#Facet Grid: general way to create different kinds of customized plots
e = sns.FacetGrid(tips, col="time",  row="smoker")
e = e.map(plt.hist, "total_bill")
```

```python
f = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
f.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

### JointGrid: analog to jointplot-method, but customizable

```python
#IMPORTANT: SEABORN
#jointgrid #multipleplot #regression #seaborn

#Joint-Grid
#general form of jointplot(), example see below
u = sns.JointGrid(x='total_bill', y='tip', data=tips)
u.plot(sns.regplot,sns.histplot)
```

```python
#example for jointplot-method
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```
